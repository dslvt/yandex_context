```python
import numpy as np
import matplotlib.pyplot as plt


def generate1():
    a = np.random.uniform(0, 1)
    b = np.random.uniform(0, 1)
    return (a * np.cos(2 * np.pi * b), a * np.sin(2 * np.pi * b))

def generate2():
    while True:
        a = np.random.uniform(-1, 1)
        b = np.random.uniform(-1, 1)
        if a ** 2 + b ** 2 > 1:
            continue
        return (a, b)
```

```python
def create_dist(generator):
    x, y = [], []
    for i in range(1000):
        a, b = generator()
        x.append(a)
        y.append(b)
    x = np.array(x)
    y = np.array(y)
    print(f'Mean x: {x.mean()}, Mean y: {y.mean()}')
    print(f'Std x: {x.std()}, Std y: {y.std()}')
    plt.scatter(x, y)
```

```python
create_dist(generate1)
```

```python
create_dist(generate2)
```
